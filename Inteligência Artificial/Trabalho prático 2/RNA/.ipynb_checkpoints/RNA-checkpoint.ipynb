# Redes Neuronais Artificiais

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import keras
from keras.models import Sequential
from sklearn.metrics import confusion_matrix

from keras.layers import Dense
from numpy import loadtxt
```

```python
# Ler o ficheiro do dataset
df =pd.read_csv('C:/Users/Pedro/Desktop/IA/TP2/Dataset/divorce_data.csv', sep = ';')
df.head() # Para ler os 5 primeiros registos
df.info()
```

```python
X = df.drop('Divorce',axis = 1)
Y = df['Divorce']
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
model = Sequential()
```

```python
#Adicionar a Input Layer e a primeira hidden layers
model.add(Dense(10, input_dim=54, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
#Compiling the ANN
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
#classifier.fit(X_train, y_train, batch_size = 10, epochs= 50)
model.fit(X_train, y_train, epochs=100, batch_size=32)
```

```python
# Evaluate the model on the test set
loss, accuracy = model.evaluate(X_test, y_test)

# Print the evaluation results
print("Test Loss:", loss)
print("Test Accuracy:", accuracy)
```

```python
df_2 =pd.read_csv('C:/Users/Pedro/Desktop/IA/TP2/Dataset/divorce_data_new.csv', sep = ';')
X_new = df_2.drop('Divorce', axis=1)
```

```python
predictions = model.predict(X_new)
predicted_classes = np.argmax(predictions, axis=1)

if predicted_classes[0] == 1:
    print("Divorce was predicted.")
else:
    print("Divorce was not predicted.")
```

```python
# Faça as previsões
y_pred = model.predict(X)

# Calcule a matriz de confusão
threshold = 0.5  # Adjust the threshold as needed
y_pred_binary = (y_pred >= threshold).astype(int)
confusion = confusion_matrix(Y, y_pred_binary)
print(confusion)
```
