# Support Vector Machines (SVM)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
# Ler o ficheiro do dataset
df = pd.read_csv('C:/Users/Pedro/Desktop/IA/TP2/Dataset/divorce_data.csv', sep = ';')
df.head() # Para ler os 5 primeiros registos
```

```python
# Dividir o dataset em variáveis preditoras (X) e variável alvo (y)
X = df.drop('Divorce', axis=1)
y = df['Divorce']

# Dividir os dados em conjunto de treinamento e conjunto de teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Criar uma instância do modelo SVM
svm_model = SVC(kernel='linear')

# Ajustar o modelo aos dados de treinamento
svm_model.fit(X_train, y_train)

# Fazer previsões utilizando o conjunto de teste
y_pred = svm_model.predict(X_test)

# Visualizar a separação de classes usando um gráfico de dispersão
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y)
plt.xlabel('Q1')
plt.ylabel('Q2')
plt.title('SVM - Separation of Classes')
plt.show()
```

```python
# Calcular as métricas
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Exibir as métricas
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
pd.DataFrame(
    confusion_matrix(y_test, y_pred),
    columns=['Predictions of Divorce', 'Predictions of No Divorce'],
    index=['Divorce', 'No Divorce']
)
```
