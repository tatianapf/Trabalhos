```python
#Criação de gráficos circulares
# Calcular o número de linhas e colunas para o layout da grade
num_questions = len(df.columns)
num_cols = 4  # Número de colunas na grade (ajuste conforme necessário)
num_rows = (num_questions - 1) // num_cols + 1

# Criar os subplots com o layout de grade desejado
fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, 25))

# Iterar por cada pergunta (assumindo colunas Q1 a Q55)
for i, column in enumerate(df.columns):
    # Calcular a posição (linha, coluna) no layout da grade
    row = i // num_cols
    col = i % num_cols

    # Obter os dados para a pergunta atual
    data = df[column]

    # Gerar o gráfico de pizza no subplot correspondente
    ax = axes[row, col]
    data.value_counts().sort_index().plot.pie(startangle=98, radius=1, labeldistance=0.5, ax=ax)
    ax.set_title(f"Pergunta {column}")
    ax.set_ylabel("")

# Remover subplots vazios
if i < (num_rows * num_cols - 1):
    for j in range(i + 1, num_rows * num_cols):
        fig.delaxes(axes.flatten()[j])

# Ajustar o espaçamento entre subplots
fig.tight_layout()

# Mostrar o gráfico
plt.show()
```

```python
# Criação de countplots

custom_palette = ["#0066ff", "#00FF00", "#0000FF"] 

fig, axes = plt.subplots(11, 5, figsize=(30, 25))
s = 0

for i in range(0, 11):
    for j in range(0, 5):
        s += 1
        if s == 55:
            break
        ax = sns.countplot(ax=axes[i, j], x=f'Q{s}', data=df, hue='Divorce', palette=custom_palette)
        ax.set_title(f'Q{s}', y=1.05)  # Posicionar o título acima do gráfico

# Ajustar a posição dos subplots para evitar sobreposição de títulos
fig.tight_layout(pad=2.0)

# Show the plot
plt.show()
```
