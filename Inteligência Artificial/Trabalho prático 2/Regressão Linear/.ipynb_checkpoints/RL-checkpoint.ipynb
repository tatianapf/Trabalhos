```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set()

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score 
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Ler o ficheiro do dataset
df =pd.read_csv('C:/Users/Pedro/Desktop/IA/TP2/Dataset/divorce_data.csv', sep = ';')
df.info()
```

```python
df.head()
```

```python
# Determinar o nÃºmero de linhas e colunas dos subplots
num_rows = (len(df.columns) - 1) // 3 + 1
num_cols = min(len(df.columns) - 1, 3)

# Criar a grade de subplots
fig, axs = plt.subplots(num_rows, num_cols, figsize=(15, 5 * num_rows))

# Iterar sobre as questÃµes Q
for i, column in enumerate(df.columns):
    if column.startswith('Q'):
        row = i // num_cols
        col = i % num_cols

        X = df[column].values.reshape(-1, 1)
        y = df['Divorce'].values

        # Criar uma instÃ¢ncia do modelo de regressÃ£o linear
        reg = LinearRegression()

        # Ajustar o modelo aos dados
        reg.fit(X, y)

        # Imprimir os coeficientes da regressÃ£o
        coef_intercept = reg.intercept_
        coef_slope = reg.coef_[0]
        coef_text = "Divorcio = {:.5} + {:.5} * {}".format(coef_intercept, coef_slope, column)

        # Realizar as previsÃµes
        f_previsoes = reg.predict(X)

        # Plotar os pontos de dados e a linha de regressÃ£o no subplot correspondente
        ax = axs[row, col]
        ax.scatter(X, y, c='red', label='Dados')
        ax.plot(X, f_previsoes, c='blue', linewidth=3, linestyle=':', label='RegressÃ£o Linear')
        ax.text(0.05, 0.95, coef_text, transform=ax.transAxes, fontsize=10, verticalalignment='top')
        ax.set_xlabel(column)
        ax.set_ylabel("($) DivÃ³rcio")
        ax.set_title("GrÃ¡fico de RegressÃ£o Linear - QuestÃ£o {}".format(column))
        ax.legend()

# Ajustar o espaÃ§o entre os subplots
plt.tight_layout()

# Exibir os subplots
plt.show()



```

```python
# Codificar variÃ¡veis categÃ³ricas usando one-hot encoding
df_encoded = pd.get_dummies(df,columns = ['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6', 'Q7', 'Q8', 'Q9', 'Q10',
           'Q11', 'Q12', 'Q13', 'Q14', 'Q15', 'Q16', 'Q17', 'Q18', 'Q19', 'Q20',
           'Q21', 'Q22', 'Q23', 'Q24', 'Q25', 'Q26', 'Q27', 'Q28', 'Q29', 'Q30',
           'Q31', 'Q32', 'Q33', 'Q34', 'Q35', 'Q36', 'Q37', 'Q38', 'Q39', 'Q40',
           'Q41', 'Q42', 'Q43', 'Q44', 'Q45', 'Q46', 'Q47', 'Q48', 'Q49', 'Q50',
           'Q51', 'Q52', 'Q53', 'Q54']
)

# Separar os dados em variÃ¡veis preditoras (X) e variÃ¡vel alvo (y)
X = df_encoded.drop('Divorce', axis=1)
y = df_encoded['Divorce']

# Dividir os dados em conjunto de treinamento e conjunto de teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=18)

# Criar uma instÃ¢ncia do modelo de regressÃ£o linear
reg = LinearRegression()

# Ajustar o modelo aos dados de treinamento
reg.fit(X_train, y_train)

# Avaliar o modelo nos dados de teste
score = reg.score(X_test, y_test)
print("Coeficiente de determinaÃ§Ã£o (RÂ²):", score)

predictions = reg.predict(X_test)
mse = mean_squared_error(y_test, predictions)
print("Erro mÃ©dio quadrÃ¡tico (MSE):", mse)
```

```python
LRmodel = LogisticRegression().fit(X_train, y_train)

y_pred = LRmodel.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

cm = confusion_matrix(y_test, y_pred)
tn = cm[0][0]
fp = cm[0][1]
fn = cm[1][0]
tp = cm[1][1]
sensitivity = tp / (fn + tp)
specificity = tn / (fp + tn)

f1 = f1_score(y_test, y_pred)

print('Accuracy:\t', round(accuracy, 2))
print('F1:\t\t', round(f1, 2))
print('Sensitivity:\t', round(sensitivity, 2))
print('Specificity:\t', round(specificity, 2))
```
